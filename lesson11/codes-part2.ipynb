```
# scatter plot
import matplotlib.pyplot as plt
import numpy as np

x=np.random.randn(1000)

y1=np.random.randn(len(x))

y2=1.2+np.exp(x)

ax1=plt.subplot(121)
plt.scatter(x,y1,color='indigo',alpha=0.3,edgecolor='white',label='no correl')
plt.xlabel('no correlation')
plt.grid(True)
plt.legend()


ax2=plt.subplot(122,sharey=ax1,sharex=ax1)
plt.scatter(x,y2,color='green',alpha=0.3,edgecolor='grey',label='correl')
plt.xlabel('strong correlation')
plt.grid(True)
plt.legend()

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

mpl.rcParams['font.size'] = 10
samples = 25

x = np.random.normal(5, 1, samples)
y = np.random.normal(3, .5, samples)

fig = plt.figure()
ax1 = fig.add_subplot(211, projection='3d')

hist, xedges, yedges = np.histogram2d(x, y, bins=10)

elements = (len(xedges) - 1) * (len(yedges) - 1)

xpos, ypos = np.meshgrid(xedges[:-1]+.25, yedges[:-1]+.25)
xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(elements)

dx = .1 * np.ones_like(zpos)
dy = dx.copy()

dz = hist.flatten()
ax1.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', alpha=0.4)
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax1.set_zlabel('Z Axis')

ax2 = fig.add_subplot(212)
ax2.scatter(x, y)
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')
plt.show()
```

```
# Get the life expectancy of several countries by bar 
import pymysql
import matplotlib.pyplot as plt

pydb=pymysql.connect('localhost','root','python3020','world')
dbcursor=pydb.cursor()

sql="select name,lifeexpectancy from country where lifeexpectancy >79"
command=dbcursor.execute(sql)
results=dbcursor.fetchall()

countrylist=[]
lifelist=[]

for row in results:
    countrylist.append(row[0])
    lifelist.append(row[1])

plt.ylim(60,100)
plt.bar(countrylist,lifelist,width=0.5,alpha=0.7)

plt.show()
```

```
#Get the sale proportion of foods by pie chart 
from pylab import *
import xlrd
from datetime import datetime
from xlrd import xldate_as_tuple

file="../data/catering_sale_all.xls"

wb=xlrd.open_workbook(filename=file)
ws=wb.sheet_by_name("Sheet1")

salelist=[]
foodlist=[]

selectdate=input("please input a date:")

for c in range(ws.ncols):
    if c>0:
        foodlist.append(ws.cell(0,c).value)
        print(ws.cell(0,c).value)

for r in range(ws.nrows):
    if r>0:
        date=datetime(*xldate_as_tuple(ws.cell(r,0).value,0))
        date=date.strftime('%Y-%m-%d')
        print(date)
        print(selectdate)
        if date==selectdate:
            for c in range(ws.ncols):
                if c>0:
                    salelist.append(ws.cell(r,c).value)
                    print(ws.cell(r,c).value)
labels=foodlist
x=salelist
explode=[]
for _ in range(len(labels)):
    explode.append(0.1)

pie(x,labels=labels,explode=explode,autopct='%1.1f%%',startangle=67)

title('The sale situation of every food on {0}'.format(selectdate))
show()
```
